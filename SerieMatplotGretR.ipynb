1) Realiza una gráfica con x contra x, x al cuadrado y x cúbica.

![grafica x](https://drive.google.com/uc?id=1Jyag35ne8a3CqaJXzB_L3mlp3dA0VtI0)

```python
x=np.linspace(0,5,100) #Crear x con valores de 0 a 5 con 100 números dentor del rango 
plt.plot(x,x, label = "Lineal", color = "blue") #Graficar funcion x
plt.plot(x,x*x,label = "Cuadratica", color = "pink") #Graficar funcion x cuadrada
plt.plot(x,x**3,label = "Cúbica", color = "purple") #Graficar x cubica
plt.legend()
plt.title("Grafica 1")  #Titulo de la grafica
plt.xlabel("Eje x")     #Titulo eje x
plt.ylabel("Eje y")     #Titulo eje y
plt.show()              #Mostrar 
```

3) A partir del siguiente diccionario despliega los datos como en la siguiente imagen. 

```python
datosFrutas = {'Manzanas' : 11, 'Naranjas' :  15, 'Bananas' : 5, 'Guayabas' : 7}
nombreFrutas = list(datosFrutas.keys())
numeroFrutas = list(datosFrutas.values())
```


![scatterplot cuadricula](https://drive.google.com/uc?id=1sugqSP4EvhYcmDZea1UQ0e-sQzoRiJZi)

NOTA/HINT: Tomen en cuenta que se usaron subplots, hay muchas formas de hacerlo, les recomiento empezar por esto:


```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize = (8, 8), sharey = False) #Crear el arreglo de 4x4 subgráficas
#ax1 será la gráfica de barras
#ax2 será la gráfica de dispersión
#ax3 será la gráfica simple de nombres x valores
#ax4 será la gráfica de pastel
ax1.bar(nombreFrutas,numeroFrutas) #bar=grafica de barras
ax1.set_title("Gráfica de Barras")

ax2.scatter(nombreFrutas,numeroFrutas) #scatter=grafica datos puntuales
ax2.set_title("Gráfica Scatter")

ax3.set_title("Ploteando") 
ax3.plot(nombreFrutas, numeroFrutas) #Plot=Grafica scatter con linea 

ax4.pie(numeroFrutas, labels=nombreFrutas,autopct='%0.2f%%')
ax4.set_title("Diagrama de Pastel") #pie=grafica de pastel
```

4) Genera un vector con números aleatorios con la shape de 1000,2 esos datos corresponderán "a" x y "y".
    Luego deberás desplegar los datos para que se vean de la siguiente manera: 
    ![scatterplot cuadricula](https://drive.google.com/uc?id=1pfrXDrwna6HEYrCw-Kf6kzBl4xptZbky)
    
    
    HINT: Es posible hacer que el color de los puntos cambien según un valor númerico, así que les sugeriría que su vector inicial fuera 1000, 3 donde la tercera columna correspondiera a 0 o 1 valores que asignen según el cuadrante al que pertenece dicha coordenada. (Lo cuál se puede hacer con una función con un par de ifs) 
    HINT: para crear un vector con números aleatorios puedes hacer una función o cimplemente implementarlo con  numpy.random.rand(numeroFilas,numeroColumnas) 

```python
puntos=np.random.rand(1000,2) #Crear matriz 1000x2, la funcion np.random.rand solo crear num aleatorios entre 0 y 1
for i in range (1000):
  x=puntos[i][0]*10 #Definir x & y
  y=puntos[i][1]*10 #Al multiplicar por 10 ampliamos el rango  
  if (x<5 and y>5) or (x>5 and y<5): #If para asignar colores 
    plt.scatter(x,y,c='#B12222')
  else:
    plt.scatter(x,y,c='#222FB1')

  
plt.show()
```

5) Este inciso es libre debes realizar una gráfica que te parezca interesante, incluso te sugiero revisar la página de matplotlib donde tienen ejemplos, debes comentar tú código (:. Los mejores 3 tendrán un punto extra en las series.

```python
r = np.arange(0, 5, 0.01) #Radio de 0 a 5
teta = 2 * np.pi * r #Angulo

ax = plt.subplot(111, projection='polar') #Graficar en polares
ax.plot(teta, r) #Valores para gráfica
ax.set_rmax(5)  #Limite en r
ax.set_rticks([1.5, 3, 4.5])  # Solo colocamos esas 4 etiquetas radiales
ax.grid(True) #"Cuadricula"

ax.set_title("Espiral en coordenadas polares")
plt.show()
```
